```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame([users,time,amount]).T
df.columns = ['users','time','amount']

def filter_time_range(ele,trange):
    if (ele>=trange[0]) and (ele<=trange[1]):
        return ele
    else:
        return np.nan
   
```

```python
def filter_time_range2(df_ele,trange):
    if (df_ele['time']>=trange[0]) and (df_ele['time']<=trange[1]):
        return df_ele['amount']
    else:
        return np.nan
```

```python
df
```

```python
sub_group_ranges = [(0,0),(20,30),(40,50),(60,60)]
for trange in sub_group_ranges:
    df[str(trange)] = df.apply(lambda x: filter_time_range2(x,trange),axis=1)
    
df
```

```python
sub_group_ranges = [(0,0),(20,30),(40,50),(60,60)]
for trange in sub_group_ranges:
    df[str(trange)] = df['time'].apply(lambda x: filter_time_range(x,trange))
    
df
```
